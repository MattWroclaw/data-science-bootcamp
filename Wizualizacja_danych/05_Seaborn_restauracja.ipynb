```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.__version__
```

```python
# dane dotyczące napiwków są wbudowane w bibliotekę seaborn
df = sns.load_dataset('tips')
df.head(10)
```

```python
df.info()
```

```python
df.describe().transpose()
```

```python
df.describe(include=['category']).transpose()
```

```python
# Tworzenie wykresów relacyjnych
# Jak wygląda relacja Rachunek::Napiwek
sns.relplot(data=df, x='total_bill', y='tip')
```

```python
# zmiana rozmiaru Label osi
sns.set(font_scale=1.2)
sns.relplot(data=df, x='total_bill', y='tip')
```

```python
# size to liczba osób w grupie; parametr też się nazywa "size"
# za pomocą 'hue' określamy kolor
sns.relplot(data=df, x='total_bill', y='tip' , hue='size')
```

```python
_=sns.relplot(data=df, x='total_bill', y='tip' , hue='size', col='time')
# "col" powinno być zmienną kategoryczną, mało rodzajów
```

```python
_=sns.relplot(data=df, x='total_bill', y='tip' , hue='size', row='time')
```

```python
_=sns.relplot(data=df, x='total_bill', y='tip' , hue='size', col='time' , row='smoker')
```

```python
# wykres Zmiennych Kategorycznych
sns.catplot(data=df, x='day', y='total_bill', hue='sex')
#
```

```python
sns.catplot(data=df, x='day', y='total_bill', kind='swarm')
```

```python
sns.catplot(data=df, x='day', y='total_bill', kind='box')
```

```python
sns.catplot(data=df, x='day', y='total_bill', kind='violin')
```

```python
sns.catplot(data=df, x='day',  kind='count')
# Piątek jest najmniej uczęszczanym dniem w restauracji / sobota najwięcej
```
