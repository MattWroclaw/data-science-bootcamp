```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('office.jpg')
img
```

```python
# wymiar obrazu
img.shape
# 426 x 640 pixels
#  3 colors
```

```python
# wyświetlenie obrazu
plt.imshow(img)
```

```python
# tworzenie siatki obrazów

fig = plt.figure(figsize=(12,10))

plt.subplot(2,2,1)
plt.imshow(mpimg.imread('office.jpg'))

plt.subplot(2,2,2)
plt.imshow(mpimg.imread('business.jpg'))

plt.subplot(2,2,3)
plt.imshow(mpimg.imread('children.jpg'))

plt.subplot(2,2,4)
plt.imshow(mpimg.imread('workplace.jpg'))

# 2 liczba wierszy
# 2 liczba kolumn
# 1 pierwszy wykres
```

```python
fig = plt.figure(figsize = (12,10))

for idx, img_name in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
  plt.subplot(220 + idx+1)
  plt.imshow(mpimg.imread(img_name))
```

```python
fig = plt.figure(figsize = (12,10))

for idx, img_name in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
  # 1 wiersz, 4 kolumny
  plt.subplot(140 + idx+1)
  plt.imshow(mpimg.imread(img_name))
```
