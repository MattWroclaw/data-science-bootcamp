```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Podstawowy obiekt to fig
fig = plt.figure()
fig.show()
```

```python
plt.plot([0,1] , [0,3])
#  Aby pozbyć się tego opisu znad wyrkesu [<matplotlib.lines.Line2D at 0x7acc26ba5c90>]
#  _ = plt.plot([0,1] , [0,3])
```

```python
import seaborn as sns
sns.set()
_ = plt.plot([0,1] , [0,3], label= 'line', color='green')
plt.xlabel('Oś x')
plt.ylabel('Oś y')
plt.title('Prosty wykres')
_ = plt.legend()
```

```python
_ = plt.plot([0,1,2] , [0,3,2], label= 'line', color='green')
plt.xlabel('Oś x')
plt.ylabel('Oś y')
plt.title('Prosty wykres')
_ = plt.legend()
```

```python
x = np.arange(-3,3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

# plt.plot(x, x_sin)
# plt.plot(x, x_cos)

plt.plot(x_cos, x_sin)

x_ss = np.sin(x_sin)
plt.plot(x, x_ss)
plt.plot(x_cos, x_ss)
```

```python
plt.subplot(121)
# 1 - ile wierszy
# 2 - ile kolumn
# 3 - konkretny wykres
plt.plot([0,1,2],[3,2,5])
plt.subplot(122)
plt.plot([0,1,2],[4,0,1] , 'r--')
```

```python
# żeby przekazać konkretny rozmiar fig, można podać w calach
fig= plt.figure(figsize=(10,5))

plt.subplot(131)
# 1 - ile wierszy
# 2 - ile kolumn
# 3 - konkretny wykres
plt.plot([0,1,2],[3,2,5])
plt.subplot(132)
plt.plot([0,1,2],[4,0,1] , 'r--')
plt.subplot(132)
plt.plot([-1,8,6],[4,0,1] , 'ro')
plt.subplot(133)
plt.plot([-1,8,6],[4,0,1] , 'ro')

# Uwaga! W PyCharm aby wyświetlić obraz należy wykonać
# fig.show()
```
