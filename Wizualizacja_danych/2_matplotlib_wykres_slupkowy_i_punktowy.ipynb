```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
matplotlib.__version__
```

```python
import seaborn as sns
sns.set()
plt.bar(x=[0,1,2,3] , height=[1.2,5.4,7.4,8.8])
```

```python
plt.bar(x=['a' , 'b', 'c', 'd'] , height=[1.2,5.4,7.4,8.8])
```

```python
plt.barh(y=['a' , 'b', 'c', 'd'] , width=[1.2,5.4,7.4,8.8])
```

```python
# generowanie ranodomowych 300 punkt√≥w

x1 = np.random.rand(300)
y1 = np.random.rand(300)

plt.scatter(x=x1,y=y1)
```

```python
# customizowanie wykresu
# s --> size kropki
x3 = np.random.rand(300)*50

plt.scatter(x=x1,y=y1, s=x3)
```

```python
# alpha --> przezroczystosc

plt.scatter(x=x1,y=y1, s=x3, alpha = 0.5)
```
