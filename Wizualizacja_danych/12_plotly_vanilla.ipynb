```python
import plotly
import plotly.graph_objs as go
plotly.__version__
```

```python
fig = go.Figure(data=go.Bar(y=[2, 3, 1, 4]) ,
                layout=go.Layout(title={'text' : 'Wykres słupkowy' }))
# można też podać title_text='Wykres słupkowy'

# zapisanie wykresu do pliku html
fig.write_html('wykres.html')

# wyświetlenie wykresu
fig.show()
```

```python
from seaborn import load_dataset
df = load_dataset('diamonds')
df.head()
```

```python
df.info()
```

```python
#  przygotowanie danych do rozkładu zmiennej 'cut'

dff = df.cut.value_counts()
dff = dff.reset_index()

# robimy to po to, żeby mieć dostępne obiekty które są w indexie
# dodatkowo zmieniamy nazwy kolumn
dff.columns = ['type', 'count']
dff
```

```python
 fig = go.Figure(
     data=go.Pie( labels=dff['type'], values=dff['count']  )
 )
 fig.show()
```

```python
fig = go.Figure(
    data=go.Pie(   labels=dff['type'], values=dff['count'], hole=0.5 )  ,
    layout=go.Layout(  title_text='Rozkład zmiennej cut', width=800      )
)
fig.show()
```
