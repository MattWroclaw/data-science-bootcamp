```python
# prompt: co robi df.reset_index() ?

# Metoda df.reset_index() resetuje indeks DataFrame'u.
# W tym przypadku, ponieważ dane pobrane z pandas_datareader.data
# mają daty jako indeks, reset_index() tworzy nową kolumnę 'Date'
# z datami i ustawia domyślny indeks numeryczny (0, 1, 2, ...).

# Przykład:
import plotly.express as px
import pandas_datareader.data as web

def fetch_financial_data(company='AMZN'):
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data()
print("DataFrame przed reset_index():")
print(df.head())

df = df.reset_index()
print("\nDataFrame po reset_index():")
print(df.head())
```

```python
px.area(df, x='Date', y='Close', log_y=True)
```
