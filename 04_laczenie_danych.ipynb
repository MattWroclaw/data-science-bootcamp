```python
import pandas as pd
pd.__version__
```

```python
def fetch_financial_data(company='AMZN'):
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

apple = fetch_financial_data('AAPL')
amazon = fetch_financial_data('AMZN')
google = fetch_financial_data('GOOGL')
uber = fetch_financial_data('UBER')
```

```python
apple.info()
#
```

Będziemy łączyć dane Apple z Uber. Jeden element wspólny to index, po których łączymy dane.

Dobra praktyka to wyświetlić kilka pierwszych wierszy


```python
uber.info()
```

```python
apple.info()
```

Żeby rozróżnić dane, zmienimy nazwy kolumn


```python
apple.columns = ['apple_' + col.lower() for col in apple.columns]
amazon.columns = ['amazon_' + col.lower() for col in amazon.columns]
google.columns = ['google_' + col.lower() for col in google.columns]
uber.columns = ['uber_' + col.lower() for col in uber.columns]
```

Sprawdzenie jak to działa na przykładzie apple

```python
apple.head(5)
```

Kontatenacja danych

```python
df =  pd.concat(objs=[apple, amazon, google, uber], axis=1)
# axis = 1 ==> względem osi Date
```

Aby zobaczyć statystyki korzystamy z .describe() wywoływanej na ramce DataFrame

```python
df.describe()
```

```python
# Aby było łatwiej patrzeć, warto przetransponować Col -> Rows
df.describe().T
```
