### Wycinanie danych

```python
price = pd.Series(data = {'Apple': 200, 'CD Projekt': 60, 'Amazon':1900, 'KGHM':np.nan})
price
```

```python
price['Apple']
```

```python
price[1]
```

### Podstawowe funkcje w Pandas

```python
#  zlicza ile jest elementów w obiekcie
price.count()
```

```python
# sprawdzenie ile razy dany el. wystąpił
price.value_counts(dropna=False)
```

```python
# sumowanie wartości w obiekcie Series
price.sum()
```

```python
price.mean()
```

```python
price.max()
```

```python
price.std()
```

Podanie statystyk zbiorczo
.describe()

```python
price.describe()
```

```python
# wybierz 2 najlepsze spółki
price.nlargest(2)
```

```python
price.nsmallest(2)
# nie zwróci nan
```

```python
# sortowanie elementów -> zwraca index
price.rank()
```

```python
# sortowanie wartości -> zwraca wartość
price.sort_values()
```

```python
# sortowanie wartości, zmiana kierunku sortowania
price.sort_values(ascending=False)
```

Metoda apply

```python
price.apply(lambda x: x*3.8)
#  to nie zmienia pierwotnego obiektu Series; nie działa tak jak shuffle w numpy gdzie obiekt jest nadpisywany
```

```python
price
```

```python
price_pln = price.apply(lambda x: x*3.8)
price_pln
```
