```python
import numpy as np
import pandas as pd
```

```python
url = ('https://storage.googleapis.com/esmartdata-courses-files/ds-bootcamp/online_retail.xlsx')
retail_raw = pd.read_excel(url)
retail_raw.head()
```

Przygotowanie danych

```python
retail = retail_raw.copy()
retail.info()
# Patrzymy ile jest kolumn, typy kolumn, jak się nazywają..
```

```python
# Patrzymy na kilka statystyk numerycznych
retail.describe().T
# Usuniemy wartości ujemne
```

Sprawdzenie braków danych

```python
retail.isnull().sum()
```

Usuwamy  baraki danych

```python
retail = retail.dropna()
retail.isnull().sum()
```

```python
retail.info()
```

```python
# usuwamy wartości ujemne z Quantity
retail = retail[retail.Quantity >= 0]
retail.info()
```

Zmieniamy CustomerID   397924 non-null  float64    na typ string

#Porównanie do języla SQL

```python
retail.CustomerID = retail.CustomerID.apply(lambda x: str(int(x)))
retail.info()
```
