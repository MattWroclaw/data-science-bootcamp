```python
import numpy as np
import pandas as pd
```

```python

import pandas_datareader.data as web


df = web.DataReader(name='META', data_source='stooq')
df.head()
```

##Zapis/Odczyt danych do pliku csv

```python
df.to_csv('meta.csv')
```

```python
df_nov = df[(df.index.month ==11) & (df.index.year == 2022)]
df_nov.head()
```

Zapisanie powyższej ramki danych do .csv

```python
df_nov.to_csv('meta_nov_2022.csv')
```

Wczytywanie danych z pliku

_UWAGA_ Kolumna Date staje się kolunmą, wcześniej służyła za index

```python
nev_df = pd.read_csv('meta_nov_2022.csv')
nev_df.head()

#
```

Zapisywanie do Excela

```python
df_nov.to_excel('meta_nov_2022.xlsx')
```

Wczytywanie danych z excela do DataFream

```python
new_df_excel = pd.read_excel('meta_nov_2022.xlsx')
new_df_excel.head()
```
